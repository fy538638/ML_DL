#### ②b. univariate imputation

```python
def univariate_impute(df, column_name):
    column = df[column_name]
    
    if pd.api.types.is_numeric_dtype(column):
        #imputed value
        mean_value = column.mean()   
        median_value = column.median()    
    
    #impute based on distribution 
        if abs(column.skew()) < 0.5:  
            imputation_value = mean_value    
        else:  
            imputation_value = median_value
        df[column_name].fillna(imputation_value, inplace=True)
    
    return df
```

```python
cols_to_impute = neededcols_G1RX203.columns[neededcols_G1RX203.isnull().mean() < 0.1]

#as mentioned above, for this column, though the proportion of data missing is 0.52% which falls into above 'cols_to_impute' category, exclude it for this moment.
excluded_col = 'PG Concentration (g/L)'  
cols_impute = [col for col in cols_to_impute if col != excluded_col]
```

#### ②b. feature selection methods ALL Together

Since the number of predictors are not large(only 15), here I didn't define the number of features I would like to have; Rather, I just let them find common ones from these 15. 

pre-test for the code:

```python
from sklearn.feature_selection import SelectKBest, f_regression
import numpy as np

# ensure y_train_direct is a NumPy array
if not isinstance(y_train_direct, np.ndarray):
    y_train_direct = np.array(y_train_direct)

# print the type and shape of y_train_direct
print(f"Type of y_train_direct: {type(y_train_direct)}")
print(f"Shape of y_train_direct: {y_train_direct.shape}")

# test the first output only (y[:, 0]) to debug step by step
def test_feature_selection(X, y, num_features):
    print(f"Testing feature selection for the first output (y[:, 0])")
    
    # Convert y[:, 0] into an explicit array to avoid slicing issues
    y_flat = y[:, 0].flatten()
    print(f"Shape of y[:, 0]: {y_flat.shape}")
    
    # Pearson Correlation feature selection
    def select_k_best_pearson(X, y, k):
        select_k_best = SelectKBest(f_regression, k=k)
        select_k_best.fit(X, y)
        return select_k_best.get_support(indices=True)
    
    # Try Pearson selection
    try:
        pearson_selected = select_k_best_pearson(X, y_flat, num_features)
        print(f"Pearson selected features: {pearson_selected}")
    except Exception as e:
        print(f"Error in Pearson selection: {e}")
    
    return pearson_selected

# run the test for the first output
selected_features = test_feature_selection(X_train_direct_scaledfs, y_train_direct, num_features=15)
print(f"Selected features for the first output: {selected_features}")
```

encapsulate: 

```python
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE

# pearson correlation
def select_k_best_pearson(X, y, k):
    select_k_best = SelectKBest(f_regression, k=k)
    select_k_best.fit(X, y)
    return select_k_best.get_support(indices=True)

# lasso
def select_lasso(X, y):
    lasso = LassoCV(cv=5).fit(X, y)
    return np.where(lasso.coef_ != 0)[0]

# random forest feature importance
def select_rf(X, y, num_features):
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X, y)
    importances = rf.feature_importances_
    indices = np.argsort(importances)[-num_features:]  # Select top 'num_features'
    return indices

# RFE
def select_rfe(X, y, num_features):
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    rfe = RFE(model, n_features_to_select=num_features)
    rfe.fit(X, y)
    return np.where(rfe.support_)[0]

# encapsulate all into one 
def get_common_features_across_outputs(X, y, num_features):
    num_outputs = y.shape[1]  # Assuming y is (samples, outputs)
    common_features = set(range(X.shape[1]))  # Start with all features
    
    for i in range(num_outputs):
        print(f"Selecting features for output {i+1}")
        
        y_flat = y[:, i].flatten()  # Flatten target for feature selection
        
        # 4 of them
        pearson_selected = select_k_best_pearson(X, y_flat, num_features)
        print(f"Pearson selected for output {i+1}: {pearson_selected}")
        
        lasso_selected = select_lasso(X, y_flat)
        print(f"Lasso selected for output {i+1}: {lasso_selected}")

        rf_selected = select_rf(X, y_flat, num_features)
        print(f"Random Forest selected for output {i+1}: {rf_selected}")
        
        rfe_selected = select_rfe(X, y_flat, num_features)
        print(f"RFE selected for output {i+1}: {rfe_selected}")
        
        # Find common features for this output
        current_common = set(pearson_selected).intersection(lasso_selected, rf_selected, rfe_selected)
        print(f"Common features for output {i+1}: {current_common}")
        
        # Intersect with overall common features
        common_features.intersection_update(current_common)
    
    return list(common_features)



# call 
common_features = get_common_features_across_outputs(X_train_direct_scaledfs, y_train_direct, num_features=15)
print(f"Common features across all outputs: {common_features}")
```

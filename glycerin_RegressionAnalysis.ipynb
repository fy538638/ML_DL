#### ②a. univariate imputation for low proportion of missing data

```python
def univariate_impute(df, column_name):
    column = df[column_name]
    
    if pd.api.types.is_numeric_dtype(column):
        #imputed value
        mean_value = column.mean()   
        median_value = column.median()    
    
    #impute based on distribution 
        if abs(column.skew()) < 0.5:  
            imputation_value = mean_value    
        else:  
            imputation_value = median_value
        df[column_name].fillna(imputation_value, inplace=True)
    
    return df


excluded_col = 'PG Concentration (g/L)'  #as we mentioned above, the missing of data is due to the non-numeric value in its calculation; but we still have to re-calculate later.
cols_to_impute = neededcols_G1RX203.columns[neededcols_G1RX203.isnull().mean() < 0.1]
cols_to_impute = [col for col in cols_to_impute if col != excluded_col]

for col in cols_to_impute:
    univariate_G1RX203 = univariate_impute(neededcols_G1RX203, col)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.multioutput import MultiOutputRegressor

gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
multioutput_gbr = MultiOutputRegressor(gbr)

multioutput_gbr.fit(X_train, y_train)

y_pred_gbr = multioutput_gbr.predict(X_test)

# Calculate metrics
mse_gbr = mean_squared_error(y_test, y_pred_gbr, multioutput='raw_values')
r2_gbr1 = r2_score(y_test, y_pred_gbr, multioutput='variance_weighted')
r2_gbr2 = r2_score(y_test, y_pred_gbr, multioutput='raw_values')

print(f'Mean Squared Error: {mse_gbr}')
print(f'R-squared: {r2_gbr1}')
print(f'R-squared: {r2_gbr2}')
```

```python
from sklearn.model_selection import GridSearchCV

# Define the model and parameters for Grid Search
gbr_trying = GradientBoostingRegressor(random_state=42)
gbr_param_grid = {
    'estimator__n_estimators': [50, 100, 200],
    'estimator__learning_rate': [0.01, 0.1, 0.2],
    'estimator__max_depth': [3, 5, 7],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4]
}

# Wrap GradientBoostingRegressor with MultiOutputRegressor
multioutput_gbr_trying = MultiOutputRegressor(gbr_trying)

# Set up Grid Search
grid_search_gbr = GridSearchCV(estimator=multioutput_gbr_trying, 
                           param_grid=gbr_param_grid, 
                           cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search_gbr.fit(X_train, y_train)
```

```python
best_params_gbr = grid_search_gbr.best_params_
best_gbr = GradientBoostingRegressor(
    n_estimators=best_params_gbr['estimator__n_estimators'],
    learning_rate=best_params_gbr['estimator__learning_rate'],
    max_depth=best_params_gbr['estimator__max_depth'],
    min_samples_split=best_params_gbr['estimator__min_samples_split'],
    min_samples_leaf=best_params_gbr['estimator__min_samples_leaf'],
    random_state=42
)
best_multioutput_gbr = MultiOutputRegressor(best_gbr)


best_multioutput_gbr.fit(X_train, y_train)

y_pred_best_gbr = best_multioutput_gbr.predict(X_test)

mse_best_gbr = mean_squared_error(y_test, y_pred_best_gbr, multioutput='raw_values')
r2_best_gbr1 = r2_score(y_test, y_pred_best_gbr, multioutput='variance_weighted')
r2_best_gbr2 = r2_score(y_test, y_pred_best_gbr, multioutput='raw_values')

print(f'Mean Squared Error: {mse_best_gbr}')
print(f'R-squared: {r2_best_gbr1}')
print(f'R-squared: {r2_best_gbr2}')
```
